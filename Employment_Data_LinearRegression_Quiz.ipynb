```python
#Features
X = df["YearsExperience"].values.reshape(-1,1)
#Target
y = df["Salary"].values
```

```python
X_train, y_train,X_test, y_test = train_test_split(X,y,test_size=0.50,random_state=10)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
lr = LinearRegression()
lr = lr.fit(X,y)
lr
```

```python
X_test = X_test.reshape(-1,1)
y_pred = lr.predict(X_test)
y_pred
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual',edgecolor = 'black')
plt.plot(X_test, y_pred, color='red', label='Predicted',linestyle =':')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
print("="*50)
mse = mean_squared_error(y_test,y_pred)
print(f"mse : {mse:.2f}")
print("="*50)
rmse = np.sqrt(mse)
print(f"RMSE : {rmse:.2f}")
print(f"On average  your predictions are off by '{rmse:.2f}' grams")
print("="*50)
mae = mean_absolute_error(y_test,y_pred)
print(f"mae : {mae:.2f}")
print(f"Your typical overall error in this code is '{mae:.2f}' grams")
print("="*50)
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"mape : {mape:.2f}")
print(f"Your predictions are only '{mape:.2f}'% wrong on average (entire data passed!)")
print("="*50)
r2 = r2_score(y_test,y_pred)
print(f"r2 : {r2:.2f}")
print(f"Your model explains 50% of the variations in brain weight")
print(f"50% of brainweight bdiff are exp by head size")
print("="*50)
```
