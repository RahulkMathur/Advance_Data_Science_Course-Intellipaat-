```python
col = list(iris.columns)
col
```

```python
# Label Encoder -
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

label_mapping ={} # creating a blank dictionary to storing object into integer-

for i in col:
  if iris[i].dtype =='object':
     iris[i] = le.fit_transform(iris[i])
     label_mapping[i] = dict(zip(le.classes_,le.transform(le.classes_)))
```

```python
for col,mapping in label_mapping.items():
  print(f'mapping for {col} : {mapping}')
```

```python
iris = iris.astype(float) # method of convertng data type to same format.
```

```python
# finding the outliers
for i in col:
  if iris[i].dtype != 'object':
    plt.boxplot(iris[i])
    plt.title(i)
    plt.show()
```
